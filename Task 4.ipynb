```python
import numpy as np
import pandas as pd
import re
import nltk
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tweets = pd.read_csv('twitter_training.csv')
tweets.head() 
```

```python
plot_size = plt.rcParams["figure.figsize"] 
print(plot_size[0])
print(plot_size[1])
plot_size[0] = 8 
plot_size[1] = 6
plt.rcParams["figure.figsize"] =plot_size
```

```python
tweets.Comment_plateform.value_counts().plot(kind='pie', autopct='%1.0f%%')
```

```python
tweets.Comment_type.value_counts().plot(kind='pie', autopct='%1.0f%%', colors=["red", "yellow", "green","pink"])
```

```python
Comment_type = tweets.groupby(['Comment_plateform', 'Comment_type']).Comment_type.count().unstack()
Comment_type.plot(kind='bar')
```

```python
import seaborn as sns
sns.barplot(x='Comment_type', y='Comment_num' , data=tweets)
```
